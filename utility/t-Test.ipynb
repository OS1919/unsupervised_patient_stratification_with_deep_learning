```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
results_frame = pd.read_csv("wandb_export_VAE.csv")
results_frame = results_frame[['mean_davis_bouldin_score_k_equals_2']]
results = results_frame.values
print ("Shape of results:", results.shape)
print("Results mean: ", np.mean(results))
#results_frame.head()
```

```python
# One-sample t-Test: H_0 <=> The mean of the sample and the mean of the population are equal
# => We test whether the observed difference is statistically significant. In this case we can reject our hypothesized population mean.
# => The null hypothesis stays the same regardless of whether the test is one-tailed or two-tailed

# Hypothesized population mean
mu = 0.68

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(results, mu, alternative='two-sided')
print("t-statistic:", t_stat)
print("p-value:", p_value)

# Setting significance level
alpha = 0.05

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis; there is a significant difference between the sample mean and the hypothesized population mean.")
else:
    print("Fail to reject the null hypothesis; there is no significant difference between the sample mean and the hypothesized population mean.")
```

```python
# Two-sample t-Test: H_0 <=> The mean between two populations is the same
# => We test whether the difference between the means of two samples (one from each population) is statistically significant.

# Perform the t-test:
# Assume that both populations have equal variances
t_stat, p_value = stats.ttest_ind(sampleA_results, sampleB_results, alternative='greater', equal_var=True)
print("p-value:", p_value)

# Interpret the results:
alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```
